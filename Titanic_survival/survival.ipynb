```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore') 
```

```python
df_train = pd.read_csv('train.csv') 
df_test = pd.read_csv('test.csv')
```

```python
y_test = pd.read_csv(r'C:\Users\LENOVO\model_deployment\test_result.csv')
```

```python
X_train = df_train.drop(columns= ['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis= 1)
X_test = df_test.drop(columns= ['PassengerId', 'Name', 'Ticket', 'Cabin'], axis= 1) 
y_train = df_train['Survived'] 
y_test = y_test['Survived']
```

```python
X_train['Age'].fillna(X_train['Age'].mean(), inplace= True)
X_train['Embarked'].fillna(X_train['Embarked'].mode()[0], inplace= True)
```

```python
X_test['Age'].fillna(X_test['Age'].mean(), inplace= True)
X_test['Fare'].fillna(X_test['Fare'].mean(), inplace= True)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Columns
col_for_le = ['Sex', 'Embarked']
col_for_scale = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# Initialize scaler
scale = StandardScaler()

# Fit + transform on training data
X_train[col_for_scale] = scale.fit_transform(X_train[col_for_scale])
# Transform test data using same scaler
X_test[col_for_scale] = scale.transform(X_test[col_for_scale])

# Label Encoding
encoders = {}
for col in col_for_le:
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
    X_test[col] = le.transform(X_test[col])  # Use same encoder for consistency
    encoders[col] = le
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize RandomForest model
rf = RandomForestClassifier(
    n_estimators=100,        # number of trees
    max_depth=7,          # grow trees fully
    min_samples_split=2,
    random_state=42,
    n_jobs=-1                # use all CPU cores
)

# Fit model on training data
rf.fit(X_train, y_train)

# Predictions
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)

# Accuracy scores
train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print(f"ðŸ”¥ Training Accuracy: {train_acc:.4f}")
print(f"ðŸ§  Testing Accuracy: {test_acc:.4f}")

# Classification Report
print("\nðŸ“Š Classification Report (Test Data):")
print(classification_report(y_test, y_test_pred))

# Optional: Confusion Matrix
print("\nðŸŒ€ Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))
```
