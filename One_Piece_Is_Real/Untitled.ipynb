```python
df = pd.read_csv('One_Piece_Episode_wise_Summary_1-1130.csv')
df.head()
```

```python
df.drop(columns= ['episode', 'character_appearances'], axis= 1, inplace= True)
df.head()
```

```python
# To Lowercase
import string
df['cln_sum'] = df['summary'].str.lower()
df['cln_ep'] = df['episode_name'].str.lower()
df.head()
```

```python
# Remove Emoji's  
import re
def remove_emoji(text): 
    return re.sub(r'[^\w\s,.!?-]', '', text)
df['cln_sum'] = df['cln_sum'].apply(remove_emoji)
df['cln_ep'] = df['cln_ep'].apply(remove_emoji)
df.head()
```

```python
# Remove URL
def remove_url(text): 
    return re.sub(r'https?://\s+\www\.\s+', '', text)
df['cln_sum'] = df['cln_sum'].apply(remove_url)
df['cln_ep'] = df['cln_ep'].apply(remove_url)
df.head()
```

```python
# Remove Stopwords 
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word.lower() not in stop_words])
df['cln_sum'] = df['cln_sum'].apply(remove_stopwords)
df['cln_ep'] = df['cln_ep'].apply(remove_stopwords)
df.head()
```

```python
# Remove Punctuations
import string
def remove_punct(text):
    return text.translate(str.maketrans('', '', string.punctuation))
df['cln_sum'] = df['cln_sum'].apply(remove_punct)
df['cln_ep'] = df['cln_ep'].apply(remove_punct)
df.head()
```

```python
# Word Tokenizer
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
df['cln_sum'] = df['cln_sum'].apply(lambda x: word_tokenize(x))
df['cln_ep'] = df['cln_ep'].apply(lambda x: word_tokenize(x))
df.head()
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
df['cln_sum'] = df['cln_sum'].apply(lambda words: [lemmatizer.lemmatize(w) for w in words])
df['cln_ep'] = df['cln_ep'].apply(lambda words: [lemmatizer.lemmatize(w) for w in words])
df.head()
```

```python
from gensim.models import Word2Vec

# flatten sentence lists
corpus = [sent for sent in df['cln_sum']]

model = Word2Vec(
    sentences=corpus,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)
```
